# **Aprendendo Básico Pandas**

```python
import pandas as pd

df = pd.read_excel('/content/base_vendas_supermercado.xlsx')
df
```

```python
#criando a coluna com o faturamento
df["Faturamento"] = df['Preco_Unitario'] * df['Quantidade']
df
```

```python
#top 3 produtos com o faturamento mais alto
top = df.groupby(['Produto'])['Faturamento'].sum().sort_values(ascending=False)
top.head(3)
```

```python
#criando o dia da semana em relação a data
df['DiaSemana'] = df['Data'].dt.day_name()
df
```

```python
#top 3 dias da semana com mais faturamento
dia_faturamento =df.groupby(['DiaSemana'])['Faturamento'].sum().sort_values(ascending=False)
dia_faturamento.head(3)
```

```python
df['Hora'] = pd.to_datetime(df['Hora'], format="%H:%M").dt.hour.astype(int)

# Cria nova coluna "Periodo"
def definir_periodo(hora):
    if 5 <= hora < 12:
        return 'Manhã'
    elif 12 <= hora < 18:
        return 'Tarde'
    elif 18 <= hora < 24:
        return 'Noite'
    else:
        return 'Madrugada'

df['Periodo'] = df['Hora'].apply(definir_periodo)

df
```

```python
df.groupby(['Cidade'])['Faturamento'].sum().sort_values(ascending = False)
```

```python
faturamento_periodo = df.groupby(['Periodo'])['Faturamento'].sum().sort_values(ascending = False)
faturamento_periodo
```

```python
df.groupby(['Filial'])['Faturamento'].sum().sort_values(ascending = False)
```

```python
df

def verifca_mes(Data):
  if Data.month == 1:
    return 'Janeiro'
  if Data.month == 2:
    return 'Fevereiro'
  if Data.month == 3:
    return 'Março'
  if Data.month == 4:
    return 'Abril'
  if Data.month == 5:
    return 'Maio'
  if Data.month == 6:
    return 'Junho'
  if Data.month == 7:
    return 'Julho'
  if Data.month == 8:
    return 'Agosto'
  if Data.month == 9:
    return 'Setembro'
  if Data.month == 10:
    return 'Outubro'
  if Data.month == 11:
    return 'Novembro'
  if Data.month == 12:
    return 'Dezembro'
df['Mes'] = df['Data'].apply(verifca_mes)
df
```

```python
filtro_mes = df.groupby(['Mes'])['Faturamento'].sum().sort_values(ascending = False)
filtro_mes
```

# **Aprendendo Seaborn**

**1. Gráfico de Dispersão (Scatterplot)**

--otimo para encontrar padroes ou correlacoes

```python
import seaborn as sns
sns.scatterplot(x= df['Faturamento'], y= df['Avaliacao'])
```

**2. Gráficos de Barras**

```python
import seaborn as sns

sns.countplot(data = df, x = 'Produto', hue = 'Cidade', dodge = True, log_scale = False, palette = "Greens")
```

**3.Gráfico de Linha**

--Usado para mostrar tendências ao longo do tempo ou sequências ordenadas.

```python
sns.lineplot(x= df['Produto'], y = df['Faturamento'])
```

```python
import seaborn as sns

# Tema visual
sns.set_theme(style="whitegrid")

# Gráfico de linha completo
sns.lineplot(
    data=df,
    x='Produto',
    y='Faturamento',
    #hue='Forma_Pagamento',         # Agrupa por categoria (se existir)
    palette='pastel',        # Paleta de cores suave
          # Adiciona marcadores nos pontos
    dashes=False,            # Linhas contínuas
    estimator='sum',         # Soma os valores por grupo
    errorbar=None,           # Sem barras de erro
    sort= True,              # Mantém a ordem original dos dados
    legend='auto'            # Exibe legenda automaticamente
)
```

**4. Histograma**

--Exibe a distribuição de uma variável numérica.

```python
sns.histplot(data=df, x='Genero', y = 'Periodo', legend='auto')
```

```python
df['Produto'].value_counts().sort_values(ascending=False)
```

**5. Boxplot**

--Mostra a distribuição, mediana, quartis e possíveis outliers.

```python
sns.boxplot(x = df['Produto'], y = df['Faturamento'])
```

**6. Violin Plot**

--Combina o boxplot com a densidade da distribuição.

```python
sns.violinplot(x = df['Produto'], y = df['Faturamento'])
```

**7. Barplot**

--Representa médias ou totais por categoria.

```python
sns.barplot(x = df['Produto'], y = df['Faturamento'])
```

**8. Jointplot**

--Combina scatter plot com histogramas marginais para duas variáveis.

```python
sns.jointplot(x = df['Faturamento'], y = df['Avaliacao'])
```

**9. Heatmap**

--Mostra correlações ou intensidades em uma matriz.

```python
import seaborn as sns
sns.heatmap(df.corr(), annot=True)

#o df corr so funciona quando todo o dataframe é somente numeros
```
