# **Aprendendo Básico Pandas**

```python
import pandas as pd

df = pd.read_excel('/content/base_vendas_supermercado.xlsx')
df
```

```python
#criando a coluna com o faturamento
df["Faturamento"] = df['Preco_Unitario'] * df['Quantidade']
df
```

```python
#top 3 produtos com o faturamento mais alto
top = df.groupby(['Produto'])['Faturamento'].sum().sort_values(ascending=False)
top.head(3)
```

```python
#criando a coluna com os dias da semana
df['DiaSemana'] = df['Data'].dt.day_name()
df
```

```python
#top 3 dias da semana com mais faturamento
dia_faturamento =df.groupby(['DiaSemana'])['Faturamento'].sum().sort_values(ascending=False)
dia_faturamento.head(3)
```

```python
df['Hora'] = pd.to_datetime(df['Hora'], format="%H:%M").dt.hour.astype(int)

# Cria nova coluna "Periodo"
def definir_periodo(hora):
    if 5 <= hora < 12:
        return 'Manhã'
    elif 12 <= hora < 18:
        return 'Tarde'
    elif 18 <= hora < 24:
        return 'Noite'
    else:
        return 'Madrugada'

df['Periodo'] = df['Hora'].apply(definir_periodo)

df
```

```python
df.groupby(['Cidade'])['Faturamento'].sum().sort_values(ascending = False)
```

```python
faturamento_periodo = df.groupby(['Periodo'])['Faturamento'].sum().sort_values(ascending = False)
faturamento_periodo
```

```python
df.groupby(['Filial'])['Faturamento'].sum().sort_values(ascending = False)
```

```python
df

def verifca_mes(Data):
  if Data.month == 1:
    return 'Janeiro'
  if Data.month == 2:
    return 'Fevereiro'
  if Data.month == 3:
    return 'Março'
  if Data.month == 4:
    return 'Abril'
  if Data.month == 5:
    return 'Maio'
  if Data.month == 6:
    return 'Junho'
  if Data.month == 7:
    return 'Julho'
  if Data.month == 8:
    return 'Agosto'
  if Data.month == 9:
    return 'Setembro'
  if Data.month == 10:
    return 'Outubro'
  if Data.month == 11:
    return 'Novembro'
  if Data.month == 12:
    return 'Dezembro'
df['Mes'] = df['Data'].apply(verifca_mes)
df
```

```python
filtro_mes = df.groupby(['Mes'])['Faturamento'].sum().sort_values(ascending = False)
filtro_mes
```
